## Python script reads a dataset, generates descriptive statistics, and a visualization using Polars
### Russian Equipment Losses (since the start of the war)
#### Data Source - Kaggle: Russia Ukraine War 2022
#### URL: https://www.kaggle.com/datasets/piterfm/2022-ukraine-russian-war?select=russia_losses_equipment.csv


First, we will begin by reading in and examining a preview of our dataset. Previewing the data set ensures it was properly imported and stored.

```python
# load required packages
import library as lib

# create a local variable and import our data
data = lib.data_read("russia_losses_equipment.csv")

# preview data to test for a successful upload
lib.data_preview(data)
```

Great! As we can see, the date set is tabular and each instance is represented by a day. Column names appear to be different kinds of equipment, but lets see all the column names before continuing.

```python
lib.data_column_names(data)
```

Excellent! Now that we know exactly all of the fields of data available to us, we need some summary statistics to gain a deeper understanding of the actual data.

```python
lib.data_summary_stats(data)
```

Summary statistics are typically very useful, but our dataset examines a change over time. Therefore, we may better understand our data if we graph it with a data visualization.

```python
lib.data_visualization(data)
```

Much better! Comparing key land vehicles (tanks, armored personnel carriers, and field artillery), we see the cumulative losses of russian ground vehicles since the war began two and half years ago. Interestingly, it appears the rate of field artillery losses has increased within the last few months whereas tank and APC loss rates are more consistent. Thanks to this data visualization our next may be to further examine the rate of loss and compare across vehicle types.
